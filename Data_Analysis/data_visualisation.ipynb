```python
# Dependencies
import pandas as pd
from pathlib import Path
import warnings
warnings.filterwarnings('ignore')
```

```python
pivot_table_df = pd.read_csv('../Outputs/pivot_table.csv')

pivot_table_df
```

```python
pivot_table_df.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 15))
sns.heatmap(pivot_table_df.set_index('Disease'), cmap='viridis', annot=False)
plt.title('Heatmap of Symptoms Across Diseases')
plt.show()
```

```python
pivot_table_df.set_index('Disease').T.plot(kind='bar', stacked=True, figsize=(15, 10))
plt.title('Stacked Bar Plot of Symptoms for Each Disease')
plt.ylabel('Symptom Severity or Occurrence')
plt.show()
```

```python
from pandas.plotting import parallel_coordinates

plt.figure(figsize=(15, 10))
parallel_coordinates(pivot_table_df, class_column='Disease', colormap='viridis')
plt.title('Parallel Coordinates Plot of Symptoms Across Diseases')
plt.show()
```

```python
# Read the dataset file with the Pandas 
dataset_df = pd.read_csv("../Resources/dataset.csv")

# Display the dataframe
dataset_df
```

```python
import matplotlib.pyplot as plt

symptom_columns = dataset_df.columns[1:]  # All symptom columns
symptom_counts = dataset_df[symptom_columns].apply(pd.Series.value_counts).sum(axis=1).sort_values()

plt.figure(figsize=(12, 8))
symptom_counts.plot(kind='barh')
plt.title('Frequency of Symptoms Across All Diseases')
plt.xlabel('Frequency')
plt.ylabel('Symptoms')
plt.show()
```

```python
import seaborn as sns
import numpy as np

selected_diseases = dataset_df['Disease'].unique()[:10]  # Select first 10 diseases for example
df_selected = dataset_df[dataset_df['Disease'].isin(selected_diseases)].set_index('Disease')
df_selected_binary = df_selected.notna().astype(int)  # Convert to binary presence/absence

plt.figure(figsize=(14, 10))
sns.heatmap(df_selected_binary.T, cmap='viridis', cbar=False)
plt.title('Heatmap of Symptoms for Selected Diseases')
plt.xlabel('Diseases')
plt.ylabel('Symptoms')
plt.show()
```
